```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df_cluster = pd.read_csv("kmodes.csv", index_col=0)
df_kmeans = pd.read_csv("kmeans.csv", index_col=0)
```

```python
df_cluster = df_cluster.sort_values(by=['kmodes_cluster'], ascending=True)
df_kmeans = df_kmeans.sort_values(by=['kmeans cluster'], ascending=True)
```

```python
fig = px.violin(df_cluster, y="Light_Conditions", box=True,
                color='kmodes_cluster',
                points='all', # can be 'outliers', or False
               )
fig.show()
```

```python
fig = px.violin(df_cluster, y="Accident_Severity", box=True,
                color='kmodes_cluster',
                points='all', # can be 'outliers', or False
               )
fig.show()
```

```python

fig = px.violin(df_cluster, y="Urban_or_Rural_Area", box=True,
                color='kmodes_cluster',
                points='all', # can be 'outliers', or False
               )
fig.show()
```

```python
fig = px.violin(df_kmeans, y="Light_Conditions", box=True,
                color='kmeans cluster',
                points='all', # can be 'outliers', or False
               )
fig.show()
```
